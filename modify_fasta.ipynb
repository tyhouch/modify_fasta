```python
import pandas as pd
import sys, re
```

```python
#opens the excel sheet as a dataframe so we can work with it in Pandas
#creates 'pango_tuple' which matches variant names to accession ID
pango = pd.read_excel('pango.xlsx')
pango = pango.rename(columns={'Unnamed: 5': 'Variant_Name', 'Unnamed: 7': 'Accession_ID'})
pango = pango[["Variant_Name", "Accession_ID"]]
pango= pango.drop(pango.index[[0,1]])
pango_tuple = list(zip(pango.Variant_Name, pango.Accession_ID))
pango_tuple
```

```python
f_old=open('muscle_align_nprot.fasta','r')
#get rid of spaces
f = open("muscle_align_nprot_no_whitespace.txt", "w+")
for line in f_old:
    f.write(re.sub("[ ]", "", line))
f= open("muscle_align_nprot_no_whitespace.txt", "r")

#read in fasta file
lines = f.readlines()
lines
```

```python
def matchID(acc):
    for var in pango_tuple:
        if var[1] == acc:
            return acc + '|' + var[0]
```

```python
#creates 'genes' an array that matches accession number with sequence
hre=re.compile('>(\S+)')
lre=re.compile('^(\S+)$')
gene={}

for line in lines:
        outh = hre.search(line)
        if outh:
                #function to convert from accession to variant name
                id = outh.group(1)
                start = id.index('EPI')
                end = id.index('.',start+1)
                accession = id[start:end]
                id = matchID(accession)
                print(id)

        else:
                outl=lre.search(line)
                if(id in gene.keys()):
                        gene[id] += outl.group(1)
                else:
                        gene[id]  =outl.group(1)
```

```python
new_fasta = open('muscle_align_nprot_data.fasta', 'w+')
for tag, seq in gene.items():
    if tag:
        new_fasta.write('>'+tag + '\n' + seq + '\n')
```
